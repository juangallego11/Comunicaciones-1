# Geometría de Números Complejos en Python

```python
import matplotlib.pyplot as plt
import numpy as np # Not really needed at this moment
import cmath

%matplotlib inline
```

```python
c = 1 + 2j # Let's start with this
print("El número complejo original Re{%d} Im{%d} y Longitud : %f" % (c.real, c.imag, abs(c)))
```

La longitud está definida como $ | a+ib | =  \sqrt{a^2 + b^2} $

### Representación Cartesiana (Plano Complejo) de un Número Complejo

```python
fig = plt.figure(figsize=(16, 8))
origin = [0], [0]
plt.quiver(*origin, c.real, c.imag, color=['r','b','g'], scale=21)
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.show()
```

### Representación Polar de Números Complejos

```python
fig = plt.figure(figsize=(16,8))
plt.polar([0, cmath.phase(c)],[0, abs(c)],marker='o')
plt.show()
```

El Angulo para $ a+ib $ está definido como $ \theta = \tan ^{ - 1}(\frac{b}{a}) $

Usando la identidad de euler podemos deshacernos de las funciones trigonométricas y escribirlas de la siguiente manera:

$ a+ib = \sqrt{a^2 + b^2} * e ^ {i*\tan ^{ - 1}(\frac{b}{a})} $

Si expresamos $ r = \sqrt{a^2 + b^2} $ y $ \theta = \tan ^{ - 1}(\frac{b}{a}) $ entonces lo anterior se puede escribir como 

$ a+ib = re ^ {i\theta} $

```python
# Proving euler identity in Python (fun activity)
import math
print ((math.e**(math.pi*1j)).real + 1)
```

```python
def rotate(number, degree):
    n = number
    return (math.e**(math.radians(degree)*1j)) * n
```

```python
c2 = rotate(c, 45)
```

```python
fig = plt.figure(figsize=(16,8))
cs = c, c2
for c in cs:
    plt.polar([0, cmath.phase(c)],[0, abs(c)],marker='o')
```

```python
def scale(number, scale):
    n = number
    return n * scale
```

```python
c3 = rotate(scale(1+2j, 2), 45)
print("Después de la operación Re{%d}, Im{%d} y longitud : %f" % (c3.real, c3.imag, abs(c3)))
```

Lo anterior es el doble del primero en longitud. $ 4.472136 = 2.236068 * 2 $

```python
fig = plt.figure(figsize=(16,8))
cs = 1+2j, c3
for c in cs:
    plt.polar([0, cmath.phase(c)],[0, abs(c)],marker='o')
```
